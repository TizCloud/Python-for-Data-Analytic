```python
# Slicing

mystring = "Python Programming"

print(mystring[1:5])
```

```python
mystring = "Welcome to Python Programming"

print('mystring[5] : ', mystring[5] )
print('mystring[2:10] : ', mystring[2:10] )
print('mystring[2:18:2] : ', mystring[2:18:2] )
print('mystring[:8] : ', mystring[:8] )
print('mystring[8:] : ', mystring[8:] )
print('mystring[::3] : ', mystring[::3] )
print('mystring[::-2] : ', mystring[::-2] )
```

```python
#loc

mydata9 = np.random.randint(10, 100, 6)
myseries9 = pd.Series(mydata9, index=[1, 4, 6, 8, 10, 12])

display (myseries9, myseries9[3:8], myseries9[3:]) ## แสดงผลเท่าที่มี
```

```python
mydata10 = np.random.randint(10, 100, 5)
myseries10 = pd.Series(mydata10, index=[1, 4, 6, 8, 10])

display (myseries10, myseries10.loc[4], myseries10.loc[3:8]) # loc ใช้ระบุข้อมูลแบบ Label Index
```

```python
mydata11 = np.random.randint(10, 100, 6)
myseries11 = pd.Series(mydata11, index=list('ABBCDE'))

display(myseries11, myseries11.loc['B':'D'])
```

```python
# Series Functions

mydata12 = [10, 20, 30, 40, 50]
myseries12 = pd.Series(mydata12)

print( 'Count : ', myseries12.count())
print( 'Mean : ', myseries12.mean())
print( 'STD : ', myseries12.std())
print( 'Max : ', myseries12.max())
print( 'Min : ', myseries12.min())

print()

print(myseries12.describe())
```

```python
mydata13 = np.random.randint(10, 100, 5)
myseries13 = pd.Series(mydata13)

display (myseries13, (myseries13 > 40).all(), (myseries13 > 40).any())
```

```python
myseries14 = pd.Series([1, 2, 4], index=list('ABC'))
myseries15 = pd.Series([8, 10], index=list('DE'))

myseriesResult = pd.concat([myseries14, myseries15])
myseriesResult
```

```python
import pandas as pd

data = { 'price': [200, 300, 250, 500],
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1],
         'discount': [8, 0, 5, 1] }

df = pd.DataFrame(data,
                  index=['p1', 'p2', 'p3', 'p4'])

display(df,
        df['price'],
        df['sales'][2],
        df['discount']['p3'],
        df.stock,
        df.stock[1],
        df.stock['p4'])
```

```python
import pandas as pd

data = { 'price': [200, 300, 250, 500],
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1],
         'discount': [8, 0, 5, 1] }

df = pd.DataFrame(data,
                  index=['p1', 'p2', 'p3', 'p4'])


display(df,
           df.at['p1', 'price'],
           df.at['p4', 'sales'],
           df.iat[0, 0],   #iat [index, column]
           df.iat[2, 3])
```

```python
import pandas as pd

data = { 'price': [200, 300, 250, 500],
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1],
         'discount': [8, 0, 5, 1] }

df = pd.DataFrame(data,
                  index=['p1', 'p2', 'p3', 'p4'])

for row in df.itertuples():
  print(f'{row.Index} {row.price} {row.sales} {row.stock}')
```

```python
import pandas as pd

data = { 'price': [200, 300, 250, 500],
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1] }

df = pd.DataFrame(data, index=['p1','p2','p3','p4'])

display((df.stock == 0).any(),
        (df.price >= 300).all(),
        (df['sales'] > 0).all())
```
